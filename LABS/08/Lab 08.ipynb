2. Create a multi-dimensional array of 3 rows and 3 columns having odd numbers from 1 to 19 including 1 and excluding 19. After that, iterate over this array to print all elements.

```python
arr = np.arange(1,19,2).reshape(3,3)
for i in np.nditer(arr):
    print(i,end=" ")
```

3. Create 3x3 matrix containing only even numbers. Multiply each element of this array with 4 and then multiply resultant matrix with 3x3 identity matrix (identity matrix should not be hardcoded).

```python
identity_matrix = np.eye(3)
even_array = np.arange(2,20,2).reshape(3,3)
even_array = 4*even_array
result = even_array * identity_matrix
print(result)
```

4. Write a NumPy program to create a structured array from given student name, height, class and their data types. Now sort by class, then height if class are equal.

```python
datatype = [('name', 'S20'), ('height', float), ('class', int)]

array = np.array([
    ('Alice', 5.5, '10'),
    ('Bob', 5.8, '10'),
    ('Charlie', 5.6, '10'),
    ('David', 5.9, '10'),
    ('Eve', 5.5, '10')
], dtype=datatype)
print(arr)

sorted_array = np.sort(array, order=['class', 'height'])
print(sorted_array)
```

5. Generate 4x4 matrix with random numbers only from (2,5,9,10) and then multiply this matrix with identity matrix. Finally add this matrix to any 4x4 matrix having only odd numbers.

```python
random_numbers = np.random.choice([2,5,9,10],size=(4,4))
identity_matrix = np.eye(4)
result = random_numbers * identity_matrix
odd_array = np.arange(1,32,2).reshape(4,4)
result = result + odd_array
print(result)
```

6. Write a NumPy program to multiply a 5x3 matrix by a 3x2.

```python
arr1 = np.arange(15).reshape(5,3)
arr2 = np.arange(6).reshape(3,2)
result = np.matmul(arr1,arr2)
print(result)
```

7. Write a NumPy program to create a random array with 1000 elements and compute the average, variance, standard deviation of the array elements. Save all the results in text file created at runtime.

```python
arr = np.random.rand(1000)
avg = arr.mean()
var = arr.var()
std = arr.std()

with open('results.txt', 'w') as f:
    f.write(f"Average: {avg}\n")
    f.write(f"Variance: {var}\n")
    f.write(f"Standard Deviation: {std}")
```
